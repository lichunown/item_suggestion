```python
def load_file(filename):
    with open(filename) as f:
        lines=[line.strip() for line in f.readlines()]
    return lines
indri_file=load_file("2009-2012.diversity.indri")
import json
with open("subtopics_intents.json") as f:
    queries_subtopics=json.load(f)
```

```python
queries_subtopics
```

```python
indri_dict={}
for line in indri_file:
    qid,_,docid,_,_,_=line.split(" ")
    if qid not in indri_dict:
        indri_dict[qid]=[]
    indri_dict[qid].append(docid)
for qid in indri_dict:
    indri_dict[qid]=indri_dict[qid][:100]
indri_dict_installed={}
import pickle
with open("full_content.pkl",'rb') as f:
    doc_dict=pickle.load(f)
for qid in indri_dict:
    query=queries_subtopics[qid]["query"]
    doc_list=indri_dict[qid]
    doc_list=[[item,doc_dict[item]] for item in doc_list]
    indri_dict_installed[qid]=[query,doc_list]
    
```

```python
with open("indri_data_installed_top100.pkl",'wb') as f:
    pickle.dump(indri_dict_installed,f)
```
